```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.title("Sports Watch Data",fontsize=20,color='k')
plt.plot(x,y,color='r',linewidth=1)
plt.xlabel('Average Pulse',fontsize=15,color='k')
plt.ylabel('Calorie Burnage',fontsize=15,color='k')
plt.grid(color ='green')
plt.show()
```

```python
#Making Bars
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])

plt.bar(x,y,color="black")
plt.show()
```

```python
###If bars to be displayed horizontally instead of vertically,use barh() function

x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])

plt.barh(x,y,color="green")
plt.show()
```

```python
#to make histogram

x = np.random.normal(100,5,2500)
plt.hist(x)
plt.show()
```

```python
#To make Pie

y = np.array([10,135,25,15,19])
plt.pie(y)
plt.show()
```

```python
'''
Add labels to pie chart with label parameter
Label parameter must be an array with one label for each wedge
'''
y = np.array([26,5,19,6,4])
mylabels = ["Apples","Cherries","Pineapple","Grapes","Berries"]
mycolors = ["Red","Pink","Yellow","Green","purple"]
plt.pie(y,labels=mylabels,colors=mycolors)
plt.show()
```

```python
y = np.array([35,25,25,15])
mylabels = ["Apples","Bananas","Cherries","Dates"]
myexplode = [0,0.95,0.8,0.2]
plt.pie(y, explode=myexplode)
plt.legend(mylabels)
plt.show()
```
